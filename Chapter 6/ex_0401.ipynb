### 1. Import thư viện

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
import os
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```


2. Tạo dữ liệu + trực quan hóa:

```python
X1, label1 = make_blobs(n_samples=100, n_features=2, centers=2, cluster_std = 5, random_state=13)
plt.scatter(X1[:,0],X1[:,1], c= label1, alpha=0.75)
plt.title('Dataset #1')
plt.show()
```

```python
X2, label2 = make_blobs(n_samples=150, n_features=2, centers=3, cluster_std = 1, random_state=26)
plt.scatter(X2[:,0],X2[:,1], c= label2, alpha=0.65 )
plt.title('Dataset #2')
plt.show()
```

### 3. Áp dụng thuật toán K-mean:

```python
kmeans = KMeans(n_clusters=5,random_state=13, n_init='auto')
kmeans.fit(X1)
myColors = {0:'red',1:'green', 2:'blue', 3:'yellow', 4:'green'}
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.75 )
plt.title('Dataset #1 : K-Means')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5,random_state=13, n_init='auto')
kmeans.fit(X1)
plt.scatter(X1[:,0],X1[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.75 )
plt.title('Dataset #1 : Chia lam 5 tap')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=13, n_init='auto')
kmeans.fit(X2)
plt.scatter(X2[:,0],X2[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.75 )
plt.title('Dataset #2 : Chia lam 5 tap')
plt.show()
```

NOTE: k means clusters have linear boundaries. The clustering result may reveal the existing structure, but this is not always guaranteed.
