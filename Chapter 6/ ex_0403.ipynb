## Coding Exercise #0403

### 1. Import thư viện

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.datasets import make_blobs, make_moons
from sklearn.cluster import KMeans
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
%matplotlib inline
```

### 2. Sinh ngẫu nhiên dữ liệu

```python
X, label = make_blobs(n_samples=500, n_features=10, centers=5, cluster_std = 5, random_state=13)
plt.scatter(X[:,0],X[:,1], c= label, alpha=0.75)
plt.title('Dataset')
plt.show()
```

### 3. Thực hiện clustering

```python
kmeans = KMeans(n_clusters=3,random_state=123, n_init='auto')
kmeans.fit(X)
myColors = {0:'red',1:'green', 2:'blue'}
plt.scatter(X[:,0],X[:,1], c= pd.Series(kmeans.labels_).apply(lambda x: myColors[x]), alpha=0.75 )
plt.title('K-Means')
plt.show()
```

### 4. Phân cụm kết hợp:

```python
agglo = AgglomerativeClustering(n_clusters=2)
agglo.fit(X)
myColors = {0:'red',1:'green'}
plt.scatter(X[:,0],X[:,1], c= pd.Series(agglo.labels_).apply(lambda x: myColors[x]), alpha=0.7 )
plt.title('Agglomerative')
plt.show()
```

```python
hehesus = linkage(X,method='single')

plt.figure(figsize=(20,5))
dendrogram(hehesus)
plt.show()
```

### 5. DBSCAN:

```python
dbscan = DBSCAN(eps=10, min_samples=5)
dbscan.fit(X)

labels = dbscan.labels_
unique_labels = sorted(set(labels))
palette = sns.color_palette("tab10", len(unique_labels))
myColors = {label: palette[i] for i, label in enumerate(unique_labels)}

plt.scatter(X[:, 0], X[:, 1], c=pd.Series(labels).apply(lambda x: myColors[x]), alpha=0.7)
plt.title('DBSCAN')
plt.show()
```
